```python
df = df.drop(columns=['customer_id','region','payment_method','number_of_profiles','watch_hours','device','favorite_genre'])
```

```python
df.head()
```

```python
## train test split
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns=['churned'])
y = df['churned']
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
## pipeline starts
numeric_cols = ['age', 'last_login_days', 'monthly_fee', 'avg_watch_time_per_day']
categorical_cols = ['gender', 'subscription_type']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
# trf1: Handle missing values
trf1 = ColumnTransformer([
    ('num_imputer', SimpleImputer(), numeric_cols),
    ('cat_imputer', SimpleImputer(strategy='most_frequent'), categorical_cols)
])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
trf2 = ColumnTransformer([
    ('cat_encoder', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [4, 5])  # these become indices of gender and subscription_type after trf1
], remainder='passthrough')
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# train the model
trf3 = DecisionTreeClassifier()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest,chi2
```

```python
# Final pipeline
pipe = Pipeline([
    ('trf1', trf1),
    ('trf2', trf2),
    ('trf3', trf3)
])

# Fit the pipeline
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
