```python
def bin_score(x):
    if x <= 3:
        return 'Low'
    elif x <= 7:
        return 'Medium'
    else:
        return 'High'

# Replace original columns with binned values
for col in ['Academic_Performance', 'Extra_Curricular_Score', 'Communication_Skills']:
    df[col] = df[col].apply(bin_score)
```

```python
df['Internship_Experience'] = df['Internship_Experience'].map({'No': 0, 'Yes': 1})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python

x = df.drop(columns=['Placement'])
y = df['Placement']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
import numpy as np

# Define transformation columns
encoder_columns = ['Academic_Performance', 'Extra_Curricular_Score', 'Communication_Skills']
scale_columns = ['IQ']

# ColumnTransformer setup
trf = ColumnTransformer([
    ('encoding',OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
, encoder_columns),
    ('scaling', MinMaxScaler(), scale_columns)
], remainder='passthrough')

# Pipeline with Linear Regression
pipe = Pipeline([
    ('preprocessing', trf),
    ('regression', LinearRegression())
])

# Encode y_train and y_test
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

# Fit the model
pipe.fit(x_train, y_train)

# Predict
y_pred = pipe.predict(x_test)

# Optional: Convert regression output to binary class
y_pred_class = np.where(y_pred >= 0.5, 1, 0)

accuracy = accuracy_score(y_test, y_pred_class)
print("Accuracy Score:", accuracy)
```
